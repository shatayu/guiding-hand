# Table Schema

| Column         | Type                      | Description                                                                 |
|----------------|---------------------------|-----------------------------------------------------------------------------|
| **uid**        | `str`                     | Unique YouTube video ID (11-character code).                                |
| **url**        | `str`                     | Input URL or canonical webpage URL.                                         |
| **title**      | `str` or `None`           | Video title.                                                                |
| **description**| `str` or `None`           | Full video description text.                                                |
| **uploader**   | `str` or `None`           | Display name of uploader.                                                   |
| **uploader_id**| `str` or `None`           | Internal YouTube uploader ID.                                               |
| **channel**    | `str` or `None`           | Display name of channel.                                                    |
| **channel_id** | `str` or `None`           | Unique channel ID.                                                          |
| **upload_date**| `str` or `None`           | Date of upload in `YYYYMMDD` format.                                        |
| **duration**   | `int` or `None`           | Duration in seconds.                                                        |
| **view_count** | `int` or `None`           | Number of views.                                                            |
| **like_count** | `int` or `None`           | Number of likes.                                                            |
| **categories** | `list[str]` or `None`     | List of category labels.                                                    |
| **tags**       | `list[str]` or `None`     | List of tags attached to video.                                             |
| **audio_path** | `str`                     | Path to the downloaded `.mp3` file.                                         |
| **whisper_model** | `str`                  | Whisper model used (`tiny`, `base`, `small`, `medium`, `large`, etc.).      |
| **language**   | `str` or `None`           | Language hint (e.g., `"en"`).                                               |
| **transcript** | `list[tuple[float, str]]` | Transcript as list of `(start_time_seconds, text)` tuples.                  |
| **thumbnails** | `list[dict]` or `None`    | List of thumbnail metadata dictionaries.                                    |
| **webpage_url**| `str` or `None`           | Canonical webpage URL of the video.                                         |
